```python
def create_bins(nums_bins_per_action=10):
    bins_cart_position = np.linspace(-4.8, 4.8, nums_bins_per_action) #4.8 -0.1
    bins_cart_velocity = np.linspace(-5,5, nums_bins_per_action)
    bins_pole_angle = np.linspace(-0.418, 0.418, nums_bins_per_action) #0.418 - 0.1
    bins_pole_velocity = np.linspace(-5,5, nums_bins_per_action)
    
    bins = np.array([bins_cart_position,
                     bins_cart_velocity,
                     bins_pole_angle,
                     bins_pole_velocity
                    ])
    
    return bins_cart_position,bins_cart_velocity,bins_pole_angle,bins_pole_velocity

#discretize observation
def discretize_observation(observations, bins):
    binned_observation = []
    
    for i, observation in enumerate(observations):
        
        discretized_observation = np.digitize(observation, bins[i])
        binned_observation.append(discretized_observation)
        
    return tuple(binned_observation)
```

```python
#action selection
def epsilon_greedy_action_selection(epsilon, q_table, discrete_state):
    
    #Exploitation
    if np.random.random() > epsilon:
        action = np.argmax(q_table[discrete_state])
    #Exploration
    else:
        action = env.action_space.sample()
        
    return action
    
#return new q
def new_q(old_q, reward, optimal_q):
    return old_q + (ALPHA*(reward + GAMMA*optimal_q- old_q))
                    
# reduce epsilon
def reduce_epsilon(epsilon,epoch):
    if BURN_IN <= epoch <= EPSILON_END:
        epsilon -= EPSILON_REDUCE
    return epsilon

#penalize for failing
def fail(done,truncated,points,reward):
    if (done or truncated) and points < 499:
        reward = -500
    return reward
```

```python
NUM_BINS = 10
BINS = create_bins(NUM_BINS)
env = gym.make('CartPole-v1')
observation,_ = env.reset()

mapped_observation = discretize_observation(observation, BINS)
print(observation)
print(mapped_observation)

EPISODES = 20000
ALPHA = 0.8 #LEARNING RATE
GAMMA = 0.95 #DISCOUNT RATE
epsilon = 1.0 #greedy-epsilon

BURN_IN = 1
EPSILON_END = 10000
EPSILON_REDUCE = 0.0001

q_table_shape = (NUM_BINS,NUM_BINS,NUM_BINS,NUM_BINS,env.action_space.n)
q_table = np.zeros(q_table_shape)
```

```python
rewards = []
log_interval = 2000


for episode in range(EPISODES):
    state, _ = env.reset()
    discretized_state = discretize_observation(state, BINS)
    done = False
    truncated = False
    points = 0
    
    #Agent plays game
    while not (done or truncated):
        
        #get action
        action = epsilon_greedy_action_selection(epsilon, q_table, discretized_state)
        #perform action
        new_state, reward, done, truncated, info = env.step(action)
        reward = fail(done,truncated, points,reward)
        
        next_discretized_state = discretize_observation(new_state, BINS)
        #old q value
        old_q_value = q_table[discretized_state + (action,)]
        
        #get next optimal q value
        optimal_q_value = np.max(q_table[next_discretized_state])
        q_table[discretized_state +(action,)] = new_q(old_q_value, reward,optimal_q_value)
        
        #track rewards
        discretized_state = next_discretized_state
        
        points += 1
        
    #Agent finish the game
    epsilon = reduce_epsilon(epsilon,episode)
    rewards.append(points)
    
    if episode % log_interval == 0 and episode != 0:
        print(np.sum(rewards[-10]))

env.close()
```

```python
indices = list(range(len(rewards)))
plt.plot(indices,rewards)
```
